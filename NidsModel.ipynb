# NIDS model

## Load Dataset

```python
import pandas as pd

df_train = pd.read_csv('Datasets/UNSW_NB15_testing-set.csv')
df_train.head()
```

```python
df_test = pd.read_csv('Datasets/UNSW_NB15_training-set.csv')
df_test.head()
```

## Data Preparation 

### Removing unnecessary variables

```python
df_train = df_train.drop(['attack_cat', 'proto', 'service', 'state', 'id', 'spkts', 'dpkts', 'sloss', 'dloss', 'ct_dst_src_ltm'], axis = 1)
df_train.head()
```

```python
df_test = df_test.drop(['attack_cat', 'proto', 'service', 'state', 'id', 'spkts', 'dpkts', 'sloss', 'dloss', 'ct_dst_src_ltm'], axis = 1)
df_test.head()
```

### Data seperation as X and Y

```python
y_train = df_train['label']
y_train.head()
```

```python
y_test = df_test['label']
y_test.head()
```

```python
x_train = df_train.drop(['label'], axis=1)
x_train.head()
```

```python
x_test = df_test.drop(['label'], axis=1)
x_test.head()
```

## Model Building

### XGBoost classification

```python
import sklearn
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(x_train, y_train)
```

### Applying model to make prediction

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
y_train_pred
```

```python
y_test_pred
```

### Evaluating the model

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_test_pred)
accuracy*100
```

### Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV

params={'max_depth':[5], 'n_estimators': [200,300,400],
            'learning_rate': [0.15,0.20], 'subsample':[0.3,0.5,0.7]}

model_2 = GridSearchCV(XGBClassifier(),params,refit=True,verbose=3)
model_2.fit(x_train, y_train)
```

```python
model_2.best_params_
```

```python
model_2.best_estimator_
```

```python
pred_2 = model_2.predict(x_test)
```

```python
score_2 = model_2.score(x_test,y_test)
print("Accuracy of 1st test model: ",score_2*100)
```
